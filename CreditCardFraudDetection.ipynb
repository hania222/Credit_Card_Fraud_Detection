#Detect and Remove Outliers

```python
def remove_outliers_iqr(data, features):
  clean_data=data.copy()
  for col in features:
    Q1=clean_data[col].quantile(0.25)
    Q3=clean_data[col].quantile(0.75)
    IQR=Q3-Q1
    lower=Q1-1.5*IQR
    upper=Q3+1.5*IQR
    #remove outliers
    clean_data=clean_data[((clean_data[col]>=lower)|(clean_data[col]<=upper))]
  return clean_data
```

```python
data_cleaned=remove_outliers_iqr(data, ['Time','Amount','Class'])
data_cleaned
```
